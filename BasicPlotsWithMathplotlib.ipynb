#Matplotlib

```python
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop)
plt.show()
```

# Scatter plot

```python
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.973]
plt.scatter(year, pop)
plt.show()
```

# Data visualization
- Many Options
  - Different plot types
  - Many Customizations
- Choice depends on
  - Data
  - Story you want to tell

```python
import matplotlib.pyplot as plt
year = [1950, 1951, 1952, 2100]
pop = [2.538, 2.57, 2.03, 10.85]
plt.plot(year, pop)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World population Projections')
plt.show()
```
